<b> Q 1. 개별 상품별 분석

```python
cnt_prd = train_past['product_name'].value_counts().reset_index()
cnt_prd.columns = ['product_name', 'frequency_count']
cnt_prd
```

```python
# 상품별 누적비중산출 
cnt_prd['cum_percent'] = 100*(cnt_prd['frequency_count'].cumsum() / cnt_prd['frequency_count'].sum())


# 상품별 비중산출
cnt_prd['percent'] = 100*(cnt_prd['frequency_count'] / cnt_prd['frequency_count'].sum())

cnt_prd
```

```python
cnt_prd_sub=cnt_prd.loc[cnt_prd['cum_percent']<50]
plt.figure(figsize=(20,10))

barplot = sns.barplot(data=cnt_prd_sub, x='product_name',y='frequency_count',hue=cnt_prd_sub['product_name'],dodge=False,palette = "hls")

for g in barplot.patches:
    barplot.annotate(format(g.get_height(), '.1f'),
                   (g.get_x() + g.get_width() / 2., g.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points')

plt.ylabel('Number of Sales', fontsize=12)
plt.xlabel('Product Name', fontsize=12)
plt.title('Total Sales By product(over 50%)', fontsize = 22)
plt.xticks(rotation='vertical')

plt.show()
```

### Day of Week

```python
def day_week(row):
    if row['week_day'] == '0':
        return 'Sun'
    elif row['week_day'] == '1':
        return 'Mon'    
    elif row['week_day'] == '2':
        return 'Tue'
    elif row['week_day'] == '3':
        return 'Wed'    
    elif row['week_day'] == '4':
        return 'Thu'    
    elif row['week_day'] == '5':
        return 'Fri'    
    return 'Sat'
    
train_past['day_week_name']= train_past.apply(day_week,axis=1)
    
train_past  
```

```python
cnt_week = train_past.groupby(['product_id','day_week_name'])['order_id'].count().reset_index(name='Sales_Number')

cnt_week_sum=cnt_week.groupby('day_week_name')['Sales_Number'].sum().reset_index(name='Sales_Sum')
cnt_week_sum['day_week_name']=pd.Categorical(cnt_week_sum['day_week_name'],categories=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
cnt_week_sum=cnt_week_sum.sort_values('day_week_name')


cnt_week_sum
```

```python
plt.figure(figsize=(20,10))

barplot = sns.barplot(data=cnt_week_sum, x='day_week_name',y='Sales_Sum',hue=cnt_week_sum['day_week_name'],dodge=False,palette = "hls")

for g in barplot.patches:
    barplot.annotate(format(g.get_height(), '.1f'),
                   (g.get_x() + g.get_width() / 2., g.get_height()),
                   ha = 'center', va = 'center',
                   xytext = (0, 9),
                   textcoords = 'offset points')

plt.ylabel('Number of order', fontsize=12)
plt.xlabel('Week Name', fontsize=12)
plt.title('Total Order  By week', fontsize = 22)
plt.xticks(rotation='vertical')

plt.show()
```

```python
order_product_cnt  = train_past.groupby('order_id').count()[['product_id']]
order_product_cnt.columns = ['product_cnt']
order_ds = train_past.merge(order_product_cnt, left_on = 'order_id',right_index =True)
```

```python


def drawWeekhour(ds, values, aggfunc=len, title=None,figsize=(20,15), cmap=None):
    weekhour_ds= ds.pivot_table(index= 'week_day', columns = 'order_hour', values=values, aggfunc = aggfunc).fillna(0)
    sns.set(style="white",palette="colorblind", font_scale=1, rc={'font.family':'NanumGothic'})
    
    plt.figure(figsize=figsize)
    f = sns.heatmap(weekhour_ds, annot=True, fmt= "1.1f", linewidths=.3, cmap=cmap)
    plt.xlabel("Hour")
    plt.ylabel("Day of Week")
    if title:
        plt.title(title,fontsize=15)
        
```

```python
drawWeekhour(order_ds, values = 'order_id',title= "Total Order Freq", aggfunc = lambda x:len(x)/1000)
```
