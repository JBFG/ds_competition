```python
train_y_past = train_x_past.iloc[:,[8]]  
train_x_past = train_x_past.drop(['product_id'], axis=1) 
train_x_past
train_y_past
```

```python
TRAIN_X = pd.concat([train_x_past, train_x_current]) 
TRAIN_X 
```

```python
TRAIN_Y = pd.concat([train_y_past, train_y]) 
TRAIN_Y 
```

```python
TEST_X = test_x_past.drop(['product_id'], axis=1) 
TEST_Y = test_x_past.iloc[:,[8]] 
```

```python
TRAIN_X_1 = TRAIN_X.drop(['order_id', 'cstno'], axis=1) 
##TRAIN_X_1['order_number','week_day','order_hour','order_interval','add_order','is_reordered'] = pd.to_numeric(TRAIN_X_1['order_number','week_day','order_hour','order_interval','add_order','is_reordered']) 
```

```python
TEST_X_1 = TEST_X.drop(['order_id', 'cstno'], axis=1) 
##TRAIN_X_1['order_number','week_day','order_hour','order_interval','add_order','is_reordered'] = pd.to_numeric(TRAIN_X_1['order_number','week_day','order_hour','order_interval','add_order','is_reordered'])
```

```python
TEST_X_1['order_number'] = pd.to_numeric(TEST_X_1['order_number']) 
TEST_X_1['week_day'] = pd.to_numeric(TEST_X_1['week_day']) 
TEST_X_1['order_hour'] = pd.to_numeric(TEST_X_1['order_hour']) 
TEST_X_1['order_interval'] = pd.to_numeric(TEST_X_1['order_interval']) 
TEST_X_1['add_order'] = pd.to_numeric(TEST_X_1['add_order']) 
TEST_X_1['is_reordered'] = pd.to_numeric(TEST_X_1['is_reordered']) 
TEST_X_1['order_interval'] = TEST_X_1['order_interval'].fillna(0) 
```
